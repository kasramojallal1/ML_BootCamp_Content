```python
import tabula

file = "https://www.cbi.ir/page/10808.aspx"

 
tables = tabula.read_pdf(file, pages = "all", multiple_tables = True)
# %%
df=tables[1].copy()
df= df.drop([0,1], axis = 0)
df
```

```python

#%%
for i in range(12,30):
    row = list(df.loc[i,:])
    row.pop()
    row.insert(0,np.nan)
    df.loc[i] =  row
df.drop('Unnamed: 0', inplace=True, axis = 1)
df.iloc[:,-1] = df.iloc[:,-1].astype(int)
df.fillna('-')
for i in range(0,df.shape[1]):
    df.iloc[:,i] = df.iloc[:,i].astype(str)
    df.iloc[:,i] = df.iloc[:,i].apply(lambda x: x.replace('/' , '.'))
    df.iloc[:,i] = df.iloc[:,i].astype(float)
# %%
df.columns=['inflation','avgYear','esfand','bahman','dey',
            'azar','aban','mehr', 'shahrivar','mordad','tir',
            'khordad', 'ordibehesht','farvardin','year']
# %%
df.reset_index(drop=True, inplace=True)
display(df.fillna(0))
#%%
```

```python
fig=px.scatter(df.dropna(), x='year', y='inflation', 
           size='avgYear', trendline='ols')

fig.show()
# %%
fig=px.bar(y=[df[col] for col in df.columns[::-1] if col !='year'], x=df['year'], 
       barmode='group',
            labels=[col for col in df.columns if col !='year'],
       title="CPI index")
# fig.update_layout(hovermode='x unified')

fig.show()
```

```python
import datetime

# %%
import pandas as pd
import pandas_datareader.data as pdr
import plotly.graph_objs as go
import plotly.offline as py

start = datetime.datetime(2020,6,19)
end = datetime.datetime(2022,7,14)
df = pdr.DataReader('BTC-USD','yahoo',start,end)
df2=pdr.DataReader('ETH-USD','yahoo',start,end)
df['return'] = df['Close']/ df['Close'].shift(1)
# %%
data = [go.Candlestick(x=df.index,
                       open=df.Open,
                       high=df.High,
                       low=df.Low,
                       close=df.Close,name='BTC/USDT') , go.Candlestick(x=df2.index,
                       open=df2.Open,
                       high=df2.High,
                       low=df2.Low,
                       close=df2.Close,name='ETH/USDT') ]

layout = go.Layout(title='Bitcoin Candlestick with Range Slider',
                   xaxis={'rangeslider':{'visible':True}})

fig = go.Figure(data=data,layout=layout)

py.iplot(fig,filename='bitcoin_candlestick')
# %%
fig=px.bar(df, x=df.index, y=df['return']-1, labels={'y':'return'})
fig.update_layout(xaxis_rangeslider_visible=True, title='Bitcoin return')
fig.show()
```

```python
def meanFinder(x):
    x= str(x)
    mylist= x.split(' ')
    if mylist[0].isdigit():
        m = (int(mylist[0]) + int(mylist[2]))/2
        return m
    elif mylist[0] == 'Younger':
        return 5
    elif mylist[0] == 'Older':
        return 64
    else: pass
    
```

```python
df['Age1stCode'] = df['Age1stCode'].apply(meanFinder)
```
